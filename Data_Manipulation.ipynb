```python
#importing the Libraries
import pandas as pd
import csv
import pickle
from sklearn.model_selection import train_test_split
import math
import numpy as np
```

```python
#path of the raw data given
raw_path = '/Users/Bhavish Pahwa/Documents/Pandas_manipulation/monojet_Zp2000.0_DM_50.0_chan3.csv'
```

```python
#opening the raw data / csv file and reading it row wise and making a list of all rows
with open(raw_path) as csvfile:
    readCSV = csv.reader(csvfile, delimiter=';')
    dates = []
    for row in readCSV:
        dates.append(row)
```

```python
#Converting the list of rows to a list which contains values related to the jets id="j"
listx=[]
for date in dates:
    for item in date:
        if 'j' in item and 'jet' not in item:
            listx.append(list(item.split(",")))
  
        
```

```python
listx
```

```python
#Defining fields of a dataframe
fields=['obj','E','pt','eta','phi']
```

```python
df=pd.DataFrame(listx,columns=fields)
```

```python
df
```

```python
#dropping the object field from the dataframe
final_df = df.drop(['obj'],axis=1)
```

```python
#Checking types of the column values of the dataframe
#print(df['E'].dtype)
final_df.dtypes
```

```python
#Converting all column values of the dataframe to float
final_df[["E", "pt","eta","phi"]] = final_df[["E", "pt","eta","phi"]].apply(pd.to_numeric)
```

```python
final_df.dtypes
```

```python
#checking of the dataframe contains any NaN / null values
final_df.isnull().sum()
```

```python
#checking if the dataframe has inf values or not
final_df.max()
```

```python
#normalizing the different columns of the dataframe according to the different scales chosen for them according to the note about the dataset given in the Zenodo entry 
def custom_normalize(final_df):
    final_df['eta']=final_df['eta']/5
    final_df['phi']=final_df['phi']/3
    final_df['E']=np.log10(final_df['E'])
    final_df['pt']=np.log10(final_df['pt'])
    return final_df
```

```python
normalized_df=custom_normalize(final_df)
```

```python
#splitting the dataset into train and validation sets with 20% as the size of validation set and 80% of train set.80/20 split after shuffling the orginal dataframe
train,test = train_test_split(normalized_df,test_size=0.2,random_state=40,shuffle=True)
```

```python
#File Paths of the new pickled files to be saved
train_final_pickle_file_path = '/Users/Bhavish Pahwa/Documents/Pandas_manipulation/train_cleaned_pickle.pkl'
test_validataion_final_pickle_file_path = '/Users/Bhavish Pahwa/Documents/Pandas_manipulation/test_cleaned_pickle.pkl'
```

```python
#Pickling according to protocol 4 so that it can work with python 3.4 and above instead of protocol 5 which works only with python 3.8 and above 
train.to_pickle(train_final_pickle_file_path,protocol=4)
test.to_pickle(test_validataion_final_pickle_file_path,protocol=4)
```

```python
#Reading pickled files/ unpickling
train_x=pd.read_pickle(train_final_pickle_file_path)
test_y=pd.read_pickle(test_validataion_final_pickle_file_path)
```
